## 1.6 Personalización del gráfico

### 1.6.1 Argumentos para personalizar el estilo y color de las líneas/gráficos:

- `color`: Permite especificar el color de la línea o marcador.
- `linestyle`: Define el estilo de la línea (sólida, discontinua, puntos, etc.).
- `marker`: Establece el marcador para cada punto de datos.
- `markersize`: Define el tamaño de los marcadores.
- `linewidth`: Establece el grosor de la línea.


````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color='red', linestyle='--', marker='o', markersize=8, linewidth=2)
plt.show()
````

### 1.6.2 Argumentos para personalizar el diseño y formato del gráfico:

- `figsize`: Tamaño de la figura en pulgadas (ancho, alto).
- `title`: Título del gráfico.
- `xlabel`, `ylabel`: Etiquetas para los ejes x e y.
- `xticks`, `yticks`: Posiciones y etiquetas de las marcas en los ejes.
- `xlim`, `ylim`: Límites de los ejes x e y.
- `grid`: Booleano que activa/desactiva la cuadrícula.



````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title('Gráfico de Líneas', fontsize=16)
plt.xlabel('Eje X', fontsize=12)
plt.ylabel('Eje Y', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.grid(True)
plt.show()
````

### 1.6.3 Argumentos para personalizar la leyenda y anotaciones:

- `label`: Etiqueta de la serie de datos para la leyenda.
- `legend`: Booleano que muestra/oculta la leyenda.
- `legend_loc`: Ubicación de la leyenda (por ejemplo, 'upper left', 'lower right').
- `legend_title`: Título de la leyenda.
- `annotate`: Anotaciones en el gráfico.


````python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 6, 9, 12, 15]

plt.plot(x, y1, label='Serie 1')
plt.plot(x, y2, label='Serie 2')
plt.legend(loc='upper left', title='Leyenda', fontsize=10)
plt.annotate('Máximo', xy=(5, 15), xytext=(4, 12),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
````

### 1.6.4 Argumentos para personalizar la apariencia de los ejes y la figura:

- `ax`: Objeto de eje en el que se dibujará el gráfico.
- `facecolor`: Color de fondo de la figura.
- `spines`: Borde de los ejes (color y grosor).
- `fontsize`: Tamaño de la fuente para los elementos del gráfico.
- `style`: Estilo predeterminado del gráfico (por ejemplo, 'ggplot', 'seaborn').


````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

fig, ax = plt.subplots(facecolor='lightgray')
ax.plot(x, y)
ax.set_title('Gráfico con Personalización de Ejes y Figura', fontsize=16)
ax.set_xlabel('Eje X', fontsize=12)
ax.set_ylabel('Eje Y', fontsize=12)
ax.tick_params(axis='both', which='major', labelsize=10)
plt.show()
````

### 1.6.5 Argumentos para la exportación y visualización del gráfico:

- `savefig`: Guardar el gráfico como un archivo de imagen.
- `dpi`: Resolución de la figura en puntos por pulgada.
- `show`: Mostrar el gráfico en pantalla o no (booleano).

````python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.savefig('grafico.png', dpi=300)
plt.show()
````

### 1.6.6 Múltiples argumentos:

````python
import matplotlib.pyplot as plt

# datos de ejemplo
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 6, 9, 12, 15]

# crear el gráfico de líneas
fig, ax = plt.subplots(figsize=(8, 6))  # tamaño de la figura
ax.plot(x, y1, color='blue', linestyle='--', marker='o', markersize=8, linewidth=2, label='Serie 1')  # línea 1
ax.plot(x, y2, color='green', linestyle='-.', marker='s', markersize=8, linewidth=2, label='Serie 2')  # línea 2

# personalización del gráfico
ax.set_title('Gráfico de Líneas Personalizado', fontsize=16)  # título
ax.set_xlabel('Eje X', fontsize=12)  # etiqueta del eje x
ax.set_ylabel('Eje Y', fontsize=12)  # etiqueta del eje y
ax.tick_params(axis='both', which='major', labelsize=10)  # tamaño de las marcas en los ejes
ax.set_xlim(0, 6)  # límites del eje x
ax.set_ylim(0, 16)  # límites del eje y
ax.grid(True)  # mostrar cuadrícula
ax.legend(loc='upper left', fontsize=10)  # leyenda
ax.annotate('Máximo', xy=(5, 15), xytext=(4, 12), arrowprops=dict(facecolor='black', shrink=0.05))  # anotación

# guardar el gráfico
plt.savefig('grafico_personalizado.png', dpi=300)

# mostrar el gráfico
plt.show()
````

# **2. Seaborn**

Seaborn es una biblioteca de visualización de datos en Python basada en Matplotlib. Está diseñada para crear gráficos atractivos y informativos de manera rápida y sencilla. Seaborn proporciona una interfaz de alto nivel para crear visualizaciones estadísticas avanzadas, lo que la convierte en una herramienta poderosa para explorar y comunicar patrones en los datos.

### 2.1 Gráfico de Distribución

```python
import seaborn as sns

sns.histplot(a, bins, kde)
```

- **Argumentos Principales:**
  - `a`: Serie de datos unidimensional.
  - `bins`: Número de contenedores para el histograma.
  - `kde`: Booleano que indica si se debe mostrar una estimación de la densidad del núcleo.
  - `stat`: Controla qué estadística se muestra en el histograma. Las opciones disponibles son:
    - `count`: Muestra el recuento de observaciones en cada bin del histograma.
    - `frequency`: Muestra la frecuencia de observaciones normalizada por el tamaño total de la muestra.
    - `density`: Muestra la densidad de probabilidad, normalizada de modo que el área total bajo el histograma sea igual a 1.
    - `probability`: Similar a `density`, pero normaliza el histograma de manera que la suma de las alturas de los bins sea 1.

- **Uso Principal:**
  - Visualizar la distribución de datos unidimensionales, incluyendo histograma y estimación de densidad del núcleo si se especifica. Esto permite entender la forma y la dispersión de los datos en una sola dimensión.

````python
# -------------solo para el ejemplo ------------------
import numpy as np
# ----------------------------------------------------

import seaborn as sns

# generar datos aleatorios
data = np.random.normal(loc=0, scale=1, size=1000)

# gráfico de distribución
sns.histplot(data, bins=30, kde=True, stat="density")
plt.show()
````

## 2.8 Personalización del Gráfico

Una de las ventajas de seaborn es su capacidad para personalizar fácilmente los gráficos. Algunas de las opciones de personalización comunes incluyen:

- **Estilo del Gráfico:** Seaborn ofrece diferentes estilos de gráficos que pueden ajustarse con la función `sns.set_style()`. Algunos estilos incluyen `whitegrid`, `darkgrid`, `ticks`, entre otros.

- **Tamaño del Gráfico:** Puedes ajustar el tamaño del gráfico utilizando los argumentos `figsize` o `size` en las funciones de seaborn o matplotlib, respectivamente.

- **Títulos y Etiquetas:** Puedes añadir títulos al gráfico y etiquetas a los ejes utilizando las funciones `plt.title()`, `plt.xlabel()` y `plt.ylabel()` de matplotlib.

- **Color del Gráfico:** Se pueden modificar los colores del gráfico utilizando los argumentos `color` y `palette` en las funciones de seaborn.

- **Líneas y Estilos:** Para gráficos como el de dispersión o de líneas, se pueden modificar los estilos de las líneas, los marcadores y otros elementos utilizando los argumentos `linestyle`, `linewidth`, `marker`, entre otros.


````python
import seaborn as sns
import matplotlib.pyplot as plt

# generar datos aleatorios
x = np.random.normal(loc=0, scale=1, size=100)
y = np.random.normal(loc=0, scale=1, size=100)

# Personalización del gráfico
sns.set_style("whitegrid")  # Establecer el estilo del gráfico
plt.figure(figsize=(8, 6))  # Establecer el tamaño del gráfico
sns.scatterplot(x=x, y=y, hue=x+y, palette='coolwarm', s=100, alpha=0.8)  # Personalización del gráfico de dispersión
plt.title('Gráfico de Dispersión Personalizado')  # Añadir un título al gráfico
plt.xlabel('Eje X')  # Añadir etiqueta al eje x
plt.ylabel('Eje Y')  # Añadir etiqueta al eje y
plt.legend(title='Suma de X y Y', loc='upper right')  # Añadir leyenda
plt.show()
````

## 2.10 Usando Seaborn y Pandas



El argumento `data` en Seaborn te permite integrar fácilmente DataFrames de Pandas en las visualizaciones, lo que facilita la creación de gráficos informativos y personalizados utilizando datos tabulares.

Para realizar esta integración se debe seguir los siguientes pasos:
1. **Cargar datos con Pandas:** Utiliza Pandas para cargar los datos desde archivos CSV, Excel u otras fuentes de datos.
2. **Crear DataFrames de Pandas:** Una vez que los datos están cargados, Pandas los organiza en DataFrames, estructuras tabulares que facilitan la manipulación y análisis de datos. Cada columna representa una variable y cada fila una observación
3. **Utilizar DataFrames con Seaborn:** Puedes pasar los DataFrames de Pandas como argumento `data` a las funciones de Seaborn. Esto permite utilizar directamente los datos de Pandas en las visualizaciones de Seaborn.
4. **Personalizar las visualizaciones:** Seaborn proporciona una variedad de funciones para crear gráficos y visualizaciones. Puedes utilizar las funciones de Seaborn para personalizar fácilmente tus visualizaciones utilizando los datos de Pandas.
5. **Manipulación de datos con Pandas:** Antes de visualizar los datos, puedes utilizar las capacidades de manipulación de datos de Pandas para filtrar, agrupar y transformar los datos según sea necesario. Esto te permite realizar análisis exploratorios de datos complejos antes de crear visualizaciones.



````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Cargar datos con Pandas:
dataset = pd.read_csv("sample_data/california_housing_train.csv")

# observar datos
dataset.head()
````

````python
# 3. Utilizar DataFrames con Seaborn:
sns.scatterplot(data=dataset, x="median_income", y="median_house_value")
````

````python
# 4. Personalizar las visualizaciones:
sns.scatterplot(data=dataset, x="median_income", y="median_house_value")
plt.title('Relación entre Ingreso Medio y Valor Medio de las Viviendas en California')
plt.xlabel('Ingreso Medio (Decenas de Miles de Dólares)')
plt.ylabel('Valor Medio de las Viviendas (Dólares)')
plt.show()
````

````python
# 5. Manipulación de datos con Pandas:
# crear 10 categorías para agrupar la variable 'housing_median_age'

dataset["housing_median_age_qcut"] = pd.qcut(dataset["housing_median_age"], q=10)
````

````python
dataset[["housing_median_age", "housing_median_age_qcut"]].sample(10)
````

````python
# agrupar los datos por la nueva variable creada y calcular el valor medio de las viviendas en cada grupo

data_agrupada = dataset.groupby('housing_median_age_qcut', as_index=False).agg({"median_house_value": "mean"})
````

````python
data_agrupada
````

````python
# graficar con los nuevos datos

plt.figure(figsize=(10, 6))
sns.barplot(x='housing_median_age_qcut', y='median_house_value', data=data_agrupada)
plt.title('Valor Medio de las Viviendas por Rango de Edad Media en California')
plt.xlabel('Rango de Edad Media de las Viviendas')
plt.ylabel('Valor Medio de las Viviendas')
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
````

# **4. Medidas de Dispersión**

## 4.1 Varianza

# **5. Detección de outliers en Pandas**

En el análisis de datos, la detección de valores atípicos (o outliers) es un paso crucial que puede afectar significativamente los resultados de nuestros estudios estadísticos. Un outlier es una observación que se desvía tanto de otras observaciones que suscita sospechas sobre si fue generada por un mecanismo diferente. La presencia de outliers puede llevar a conclusiones erróneas y afectar la precisión de los modelos predictivos.

## Detección de Outliers con el Método IQR en Pandas
El proceso para detectar outliers con el método IQR en Pandas es relativamente directo:

1. **Calcular el IQR:** Primero, determinamos el IQR de la variable de interés.
2. **Establecer los Límites:** A continuación, definimos los límites inferior y superior para los outliers. Usualmente, los outliers se consideran aquellos valores que están por debajo de `Q1 - 1.5 * IQR` o por encima de `Q3 + 1.5 * IQR`.
3. **Filtrar los Outliers:** Finalmente, podemos filtrar los datos para excluir o atender específicamente a los outliers identificados según estos límites.

````python
import pandas as pd
import numpy as np
import seaborn as sns

# cargar datos
dataset = sns.load_dataset("taxis")

# calcular cuartiles
q1 = np.quantile(dataset["fare"], 0.25)
q3 = np.quantile(dataset["fare"], 0.75)

# obtener IQR
iqr = q3 - q1

# generar limites
lower_lim = q1 - (1.5 * iqr)
upper_lim = q3 + (1.5 * iqr)

# filtrar por outliers
dataset_no_outliers = dataset[
  (dataset["fare"] > lower_lim) &
  (dataset["fare"] < upper_lim)
]
````
