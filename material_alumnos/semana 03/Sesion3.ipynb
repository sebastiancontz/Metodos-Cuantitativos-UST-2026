# **Crear Dataframes**

Puedes crear un DataFrame de varias maneras, pero una de las más comunes es a partir de un diccionario de listas o arrays NumPy

## **Desde Diccionarios**

````python
# Se puede crear directamente desde un Diccionario de Listas o Arrays:
data = {'Nombre': ['Juan', 'Ana', 'Luis'],
        'Edad': [25, 30, 22],
        'Ciudad': ['Santiago', 'La Serena', 'Temuco']}

df = pd.DataFrame(data)
print(df)
````

````python
# hay un método específico para diccionarios (pd.DataFrame.from_dict) que funciona similar al normal, pero tiene mas opciones
data = {'Nombre': ['Juan', 'Ana', 'Luis'],
        'Edad': [25, 30, 22],
        'Ciudad': ['Santiago', 'La Serena', 'Temuco']}


df = pd.DataFrame.from_dict(data)
print(df)
````

````python
# Acá pd.DataFrame.from_dict permite que la orientación cambie y que en vez de tener
# índices numéricos tengamos los nombres de las variables

data = {'Nombre': ['Juan', 'Ana', 'Luis'],
        'Edad': [25, 30, 22],
        'Ciudad': ['Madrid', 'Barcelona', 'Valencia']}

# Usar el constructor from_dict con orientación 'index'
df = pd.DataFrame.from_dict(data, orient='index', columns=['Persona1', 'Persona2', 'Persona3'])
# acá cambiamos la orientación con orient y le asignamos nombres a las columnas (personas)
print(df)
````

## **Desde Listas**

````python
nombres = ['Juan', 'Ana', 'Luis']
edades = [25, 30, 22]

# Crear un DataFrame desde dos listas
df = pd.DataFrame({'Edad': edades, 'Nombre': nombres}) # Le pasamos un dicionario con el nombre de la columna:valor
print(df)
````

````python
data_list_of_lists = [['Juan', 25, 'Santiago'],
                      ['Ana', 30, 'La Serena'],
                      ['Luis', 22, 'Temuco']]

df = pd.DataFrame(data_list_of_lists, columns=['Nombre', 'Edad', 'Ciudad'])
print(df)
````

##**Conteo de elementos diferentes en una columna**

El método `value_counts()` en Pandas se utiliza para contar la ocurrencia de valores únicos en una Serie. Este método proporciona información sobre la distribución de valores y la frecuencia de cada valor único en una columna.

```python
serie.value_counts(normalize=False, sort=True, ascending=False, bins=None, dropna=True)

```

* **normalize:** Si es `True`, devuelve las frecuencias relativas en lugar de las frecuencias absolutas. Por defecto, es `False`.
* **sort:** Si es `True`, ordena los resultados por frecuencia. Por defecto, es True.
* **ascending:** Si es `True`, ordena las frecuencias en orden ascendente. Por defecto, es `False`.
* **bins:** Solo para datos numéricos. Divide los valores en intervalos y cuenta la frecuencia en cada intervalo.
*  **dropna:** Si es `False`, incluye los valores nulos en el resultado. Por defecto, es `True`.

##**Eliminar registros nulos dropna()**

El método `dropna() `en Pandas se utiliza para eliminar filas o columnas que contienen valores nulos (**NaN**) de un DataFrame. Este método ofrece flexibilidad para adaptarse a diferentes necesidades de limpieza de datos.

```python
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

```

* **axis:** Especifica si se eliminarán **filas (axis=0)** o **columnas (axis=1)**. Por defecto, es 0 (filas).
* **how:** Controla cuándo se eliminará una fila o columna. Puede ser '**any**' (si hay al menos un NaN) o '**all**' (solo si todos son NaN). Por defecto, es 'any'.
* **thresh:** Número mínimo de valores no nulos necesarios para que una fila o columna no sea eliminada.
* **subset:** Lista de etiquetas de índice o nombres de columnas en los cuales buscar valores nulos.
* **inplace:** Si es `True`, realiza la eliminación en el lugar y modifica el DataFrame original. Por defecto, es `False`

````python
print(df.shape) # Tamaño original (891, 12)
````

````python
df2 = df.dropna()
print(df2.shape)
````

````python
df3 = df.dropna(subset=['Age'])  # Sólo eliminamos donde Age es nulo (714, 12)
print(df3.shape)
````

##**Reemplazando los registros nulos fillna()**

El método `.fillna()` en Pandas se utiliza para rellenar los valores nulos (`NaN`) en un DataFrame o una Serie con valores específicos o mediante ciertos métodos de interpolación. La idea principal es no perder registros valiosos.

```python
df.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None)
```

<h3><b>Algunos parámetros:</h3></b>

* **value:** Valor escalar, diccionario o Series que indica qué valor usar para rellenar NaN. Por defecto, es None.
* **method:** Método de rellenado, como **'ffill' (forward fill)** o '**bfill' (backward fill)**. Por defecto, es None.
* **axis:** Eje a lo largo del cual se realizará el rellenado (0 para filas, 1 para columnas). Por defecto, es None (rellenado en ambos ejes).
* **inplace:** Si es `True`, realiza el rellenado en el lugar y devuelve None. Por defecto, es `False`.

# **Eliminar columnas**

En pandas, el concepto de crear columnas en un DataFrame se refiere a eliminar una o más series de datos del DataFrame existente.

El método `drop` se utiliza para eliminar columnas o filas de un DataFrame. Cuando se usa para eliminar columnas, se especifica el parámetro `axis=1`.

```python
dataframe.drop(labels, axis=1, inplace=False)
```

* **labels:** Puede ser el nombre de una columna o una lista de nombres de columnas que deseas eliminar.

* **axis:** Debe ser establecido en 1 para indicar que estás eliminando columnas. Si quisieras eliminar filas, usarías axis=0.

* **inplace:** Un parámetro opcional que, si se establece en `True`, realizará la modificación en el lugar (es decir, **modificará el DataFrame original**). Si se establece en `False` **(valor predeterminado), devolverá un nuevo DataFrame con las columnas eliminadas**.

* **errors:** Determina que sucede si se le pasa una columna que no existe en el dataframe. Por defecto el valor es **'raise'**, generando error si no encuentra la columna y cayendo el programa. Pero se le puede pasar el valor **'ignore'** que no generar un error (evita que caiga el programa).

# **Agrupando Data**

La agrupación de datos con Pandas es una técnica poderosa para analizar conjuntos de datos mediante la segmentación de datos en grupos basados en determinados criterios. Esto implica:

- **Splitting** (División): Separar los datos en grupos basados en un criterio especificado.
- **Applying** (Aplicación): Aplicar una función a cada grupo independientemente.
- **Combining** (Combinación): Combinar los resultados en una estructura de datos.

* Pandas ofrece la función `groupby()` que permite realizar estas operaciones de manera eficiente. Se utiliza para dividir los datos en grupos basados en una columna específica o una serie de columnas. Luego, se pueden aplicar operaciones (como sumas, promedios, conteos, etc.) a estos grupos.


```python
df.groupby('VarCategorica')

df.groupby(['VarCategorica1', 'VarCategorica2']) # se pueden agrupar por varias
```




**Principales funciones**

* **groupby()**: Método para crear un objeto GroupBy.
* **agg()**: Para realizar agregaciones sobre grupos (por ejemplo, sumas, promedios, conteos).
* **apply()**: Para aplicar funciones personalizadas a grupos de datos.
* **transform()**: Para realizar transformaciones en los datos de cada grupo.
* **size()**: Para obtener el tamaño de cada grupo.
* **first(), last():** Para obtener el primer o último elemento de cada grupo.
* **describe()**: Para obtener estadísticas descriptivas por grupo.
* **filter()**: Para filtrar datos en función de propiedades de los grupos, retornando solo los registros en los que ciertos grupos cumplen la condición.

### **Sacando las estadisticas con agg()**

La función agg en pandas se utiliza para aplicar funciones de agregación a grupos de datos. Permite realizar operaciones más complejas y personalizadas que las funciones predefinidas de agregación como mean, sum, etc.

**Sintaxis:**

```python
resultado_agg = df.groupby('VarCategorica').agg({'ColumnaNumerica': [lista funciones ]})

####Ejemplo funciones predefinidas ####
resultado_agg = df.groupby('VarCategorica').agg({'ColumnaNumerica': ['mean', 'median', 'sum', 'min', 'max']})

####Ejemplo funciones personalizadas ####
def mifunc():
  ###hace algun calculo

resultado_agg = df.groupby('VarCategorica').agg({'ColumnaNumerica': ['mean', mifunc]})

```



````python
agregaciones_multiples = grouped.agg({'Fare': ['mean', 'min', 'max']})
agregaciones_multiples
````

````python
# Forma alternativa de llamar la función agg
agregaciones_multiples = grouped['Fare'].agg([min, max, sum])
agregaciones_multiples
````

````python
# Aplicar múltiples funciones de agregación a la vez
# Primeros nos quedamos sin los valores 0 y luego agrupamos con groupby
agregaciones_multiples = df[df["Fare"]>0].groupby('Pclass').agg({'Fare': ['mean', 'min', 'max']})
agregaciones_multiples
````

````python
def rango(x):
  """
    Calcula el rango estadístico de un conjunto de datos (diferencia entre el
    valor máximo y el valor mínimo en un conjunto de datos).

    Parámetros:
    - x (pandas.Series): Serie de pandas que representa una columna de datos.

    Retorna:
    - float: El rango estadístico, es decir, la diferencia entre el valor máximo y el valor mínimo en x.
  """
  return x.max() - x.min()
````

````python
agregaciones_multiples_prersonalizadas = grouped.agg({'Fare': ['mean', 'median', 'sum', rango]})
agregaciones_multiples_prersonalizadas
````

###**Transformando los grupos**
El método `transform()` en groupby se utiliza para realizar operaciones de transformación en cada grupo de datos y devolver un objeto con la misma forma que el objeto original.

**Sintaxis**:

`DataFrame.groupby('clave').transform(func, *args, **kwargs)`

* **func:** Función, función anónima (lambda), o nombre de función que se aplicará a cada grupo.

### **Ejemplo de uso: Llenando los valores faltantes**

Puedes utilizar transform para llenar los valores faltantes en cada grupo con la media del grupo


````python
df["Age"].isna().sum()
````

````python
df['Age_faltante'] = df.groupby('Pclass')['Age'].transform(lambda x: x.fillna(x.mean()))
````

````python
df["Age_faltante"].isna().sum()
````

### **Filtrando con filter()**

El método `filter()` en `groupby` permite filtrar grupos de datos según una condición y devolver un nuevo DataFrame que incluye solo los grupos que cumplen con esa condición.

```python
DataFrame.groupby('clave').filter(func, *args, **kwargs)

```
* **func:** Función, función anónima (**lambda**), o nombre de función que devuelve `True` o `False`.



````python
# Filtrando los grupos cuyo promedio de valor de ticket es mayor a 20
filtrando_grupos =  df[df["Fare"]>0].groupby('Pclass').filter(lambda x: x['Fare'].mean()> 20)
filtrando_grupos.head(2)
````

````python
filtrando_grupos["Pclass"].value_counts()  # Eliminando una clase (3)
````
