# 1 Introducción a los Tipos de Datos en Python

En esta sesión, exploraremos los tipos de datos fundamentales en Python, esenciales para manipular información y realizar operaciones en el lenguaje.


Comprender los tipos de datos es fundamental para escribir código eficiente y evitar errores en el procesamiento de la información.

En la sesión anterior, nos enfocamos en tres categorías principales:

1. **Tipos Numéricos (`int`, `float`, `complex`)**: Representan valores numéricos y permiten operaciones matemáticas.

3. **Tipos Booleanos (`bool`)**: Representación de valores lógicos (`True` y `False`), operadores lógicos y comparaciones.


## 1.1 El valor `None` en Python
`None` es un objeto especial en Python que representa la ausencia de un valor o un valor nulo. Se utiliza para indicar que una variable no tiene un valor asignado.

```python
variable = None
print(variable)
```

```python
if variable is None:
    print("La variable no tiene un valor asignado")
```

# 2 Introducción a las Estructuras de Datos en Python

Las **estructuras de datos** son formas organizadas de almacenar, gestionar y manipular información en un programa. En Python, existen varias estructuras que permiten trabajar con diferentes tipos de datos de manera eficiente.


## 2.1 ¿Por qué son importantes las estructuras de datos?
Las estructuras de datos son fundamentales en programación porque:
- Permiten organizar y acceder a la información de manera eficiente.
- Optimizan el uso de memoria y procesamiento.
- Son clave en la resolución de problemas computacionales y algoritmos.


## 2.2 Tipos de Estructuras de Datos en Python
Python proporciona varias estructuras de datos incorporadas (*built-in*), cada una con características y usos específicos:


### Listas (`list`)
Las listas son estructuras **ordenadas** y **mutables**, lo que significa que podemos modificar sus elementos después de su creación.


```python
mi_lista = [10, 20, 30, "Python"]
mi_lista.append(40)
print(mi_lista)
```

```python
print(mi_lista[-1])
```

```python
mi_lista[-1] = 40
```

```python
print(mi_lista)
```

### Tuplas (`tuple`)

Las tuplas son similares a las listas, pero inmutables (no pueden modificarse después de su creación).

```python
mi_tupla = (1, 2, 3, "Python")
print(mi_tupla[0])
```

```python
print(mi_tupla[-1])
```

```python
# Error: las tuplas son inmutables
mi_tupla[-1] = 40
```

### Diccionarios (`dict`)

Los diccionarios almacenan datos en pares clave-valor, permitiendo acceso rápido a los valores a través de sus claves.

```python
mi_diccionario = {
    "nombre": "Ana",
    "edad": 25,
    "ciudad": "Santiago"
}

print(mi_diccionario["nombre"])
```

### Conjuntos (`set`)

Los conjuntos son estructuras desordenadas que almacenan elementos únicos.

```python
mi_conjunto = {1, 2, 3, 3, 4}
print(mi_conjunto)
```

## 2.3 NumPy

**NumPy** (*Numerical Python*) es una biblioteca que proporciona soporte para **arrays multidimensionales** y funciones matemáticas de alto rendimiento. Es utilizada principalmente para realizar operaciones numéricas eficientes, ya que está optimizada para cálculos vectorizados.


### Características principales de NumPy
- Soporte para **arrays multidimensionales (`ndarray`)**  
- Operaciones matemáticas y algebraicas optimizadas  
- Integración con otras bibliotecas científicas como `SciPy`, `Matplotlib` y `Pandas`  
- Mayor eficiencia en memoria y velocidad en comparación con listas de Python  

### Vectores en NumPy

Un vector es un array unidimensional, usado comúnmente en álgebra lineal y machine learning.

```python
# importamos la librería de NumPy con el alias 'np'
import numpy as np

vector = np.array([5, 10, 15])
print(vector)
```

#### Operaciones comunes con vectores

```python
# multiplicar por un escalar
print(vector * 2)
```

```python
# obtener la media del vector
print(np.mean(vector))
```

```python
# obtener la suma acumulada del vector
print(np.cumsum(vector))
```

```python
# obtener la suma total del vector
np.sum(vector)
```

```python
# operaciones entre vectores
vector_1 = np.array([3, 4, 5])
vector_2 = np.array([7, 8, 9])
```

```python
# suma
print(vector_1 + vector_2)
```

```python
# resta
print(vector_1 - vector_2)
```

```python
# multiplicación
print(vector_1 * vector_2)
```

```python
# división
print(vector_1 / vector_2)
```

### Matrices en NumPy

Una matriz es un array bidimensional, representado como una tabla con filas y columnas.

```python
matriz = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print(matriz)
```

```python
# obtener la dimensión (fila, columna)
print(matriz.shape)
```

#### Operaciones comunes con matrices

```python
# transponer la matriz
print(matriz.T)
```

```python
# producto punto
print(np.dot(matriz, matriz.T))
```

### Diferencias entre listas y arrays de NumPy

| **Característica**       | **Lista de Python**    | **Array de NumPy** |
|-------------------------|----------------------|--------------------|
| **Tipo de datos**      | Permite distintos tipos | Requiere tipos homogéneos |
| **Eficiencia en memoria** | Menos eficiente         | Más eficiente |
| **Soporte para cálculos** | Operaciones manuales    | Operaciones vectorizadas rápidas |

## 2.4 Pandas

**Pandas** es una biblioteca diseñada para manejar y analizar datos estructurados en **Series** y **DataFrames**. Se basa en NumPy y permite trabajar con grandes volúmenes de datos de manera eficiente.

### Series en Pandas

Una **Serie** en Pandas es una estructura de **una sola columna** con un índice asociado. Es similar a un **array de NumPy**, pero con etiquetas para cada valor.

```python
# importamos la librería Pandas con el alias 'pd'
import pandas as pd

# crear una Serie en Pandas
serie = pd.Series(
    data=[10, 20, 30], # valores
    index=["2000", "2001", "2002"] # indices
)
print(serie)
```

#### Operaciones comunes en Series

```python
# acceder a un valor a través del índice
print(serie["2000"])
```

```python
# calcular el promedio
print(serie.mean())
```

```python
# calcular la suma acumulada
print(serie.cumsum())
```

```python
# calcular la suma de la serie
print(serie.sum())
```

## DataFrames en Pandas

Un **DataFrame** es una estructura tabular de **múltiples columnas**, similar a una hoja de cálculo o tabla de base de datos.

```python
datos = {
    "Nombre": ["Ana", "Carlos", "Sofía"],
    "Edad": [25, 30, 22]
}
df = pd.DataFrame(datos)
print(df)
```

#### Operaciones comunes en DataFrames

```python
# primeras filas (por defecto 5)
print(df.head())
```

```python
# primeras n filas
print(df.head(1))
```

```python
# ultimas n filas
print(df.tail(1))
```

```python
# estadísticas de columnas numéricas
print(df.describe())
```

```python
# promedio de la columna "Edad"
print(df["Edad"].mean())
```

## Diferencias entre NumPy y Pandas

| **Característica**       | **Lista de Python**    | **Array de NumPy** |
|-------------------------|----------------------|--------------------|

| **Eficiencia en memoria** | Menos eficiente         | Más eficiente |

-	🏎 **NumPy** es ideal para **cálculos numéricos** y **procesamiento de datos en matrices**.
-	📊 **Pandas** es mejor cuando trabajas con **tablas de datos (como hojas de cálculo o bases de datos)**.

# 3 Estructuras de Control de Flujo

- Python permite controlar la ejecución de un programa a través de **estructuras de control de flujo**, como condicionales (`if`, `else`) y bucles (`for`, `while`).
- Las estructuras de control permiten modificar el flujo de ejecución de un programa con **condicionales** y **bucles**.



## 3.1 Condicionales (`if`, `elif`, `else`)
Las condicionales permiten ejecutar código dependiendo de una condición.

```python
edad = 18

if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
```

### Uso de `elif` (else if)

Podemos agregar más condiciones con elif:

```python
nota = 55

if nota >= 65:
    print("Excelente")
elif nota >= 40:
    print("Aprobado")
else:
    print("Reprobado")
```

📌 Nota: Python usa **indentación** para definir bloques de código.

```python
edad = 35

if edad <=2:
    print("Bebe")
elif edad <= 14:
    print("Niño")
elif edad <= 20:
    print("Adolecente")
elif edad <= 40:
    print("Adulto Joven")
elif edad <= 60:
    print("Adulto")
else:
    print("Adulto Mayor")
```

## 3.2 Bucles en Python

Los bucles permiten repetir instrucciones múltiples veces.

### Bucle `for`

Se usa para iterar sobre secuencias como listas, tuplas o rangos.

```python
frutas = ["Manzana", "Banana", "Cereza"]

# por cada elemento dentro de la lista frutas
for fruta in frutas:
    print(fruta)
```

```python
# el iterador puede tener el nombre que determine más conveniente
for elemento in frutas:
    print(elemento)
```

#### Uso de `range()` en `for`

```python
# iterara de 0 a 4 (el método range no incluye el punto de corte)
for i in range(5):
    print(i)
```

```python
# empezar en 1 e iterar hasta 4
for i in range(1, 5):
    print(i)
```

```python
# empezar en 2 e iterar hasta 10, de 2 en 2
for i in range(2, 11, 2):
    print(i)
```

### Bucle `while`

Ejecuta el bloque de código mientras la condición sea True.

```python
contador = 0

while contador < 3:
    print(f"Iteración {contador}")
    contador += 1
```

#### `break` y `continue` en bucles
-	`break`: Detiene el bucle.
-	`continue`: Salta a la siguiente iteración.

```python
for i in range(5):
    if i == 3:
        break  # detiene el bucle cuando i es 3
    print(i)
```

```python
for i in range(5):
    if i == 2:
        continue  # salta el valor 2 y sigue con la siguiente iteración
    print(i)
```

# 4 Creación de Funciones en Python

- Las funciones en Python permiten modularizar el código, haciéndolo más **organizado, reutilizable y eficiente**.
- En esta sección, exploraremos en profundidad cómo definir funciones, utilizar parámetros, valores de retorno y otros conceptos avanzados.


## 4.1 ¿Qué es una función?
Una **función** es un bloque de código reutilizable que se ejecuta solo cuando se llama. Se pueden pasar datos a la función como **parámetros** y devolver valores usando `return`.


## 4.2 Definir una Función con `def`

Las funciones en Python se definen con la palabra clave `def`, seguida del **nombre de la función** y paréntesis `()`, donde pueden ir los **parámetros**.


```python
def saludar():
    print("¡Hola, bienvenido a Python!")
```

```python
# llamar la función
saludar()
```

## 4.3 Parámetros en Funciones

Los **parámetros** permiten que una función reciba valores para procesarlos dentro de su bloque de código.

```python
# @title Función con Parámetros
def saludar(nombre):
    print(f"Hola, {nombre}")

saludar("Carlos")
saludar("Ana")
```

Las funciones también nos permite usar valores por defecto, en caso que no se proporcione un valor.

```python
# @title Parámetros con valores predeterminados

def saludar(nombre="Invitado"):
    print(f"Hola, {nombre}")

saludar("Elena")
saludar()
```

```python
# @title Función con múltiples Parámetros

def sumar(a, b):
    return a + b  # Devuelve la suma de a y b

resultado = sumar(4, 7)
print(resultado)
```

## 4.4 Retorno de Valores con `return`

Una función puede devolver un resultado con la palabra clave return.

```python
def cuadrado(numero):
    return numero ** 2

resultado = cuadrado(5)
print(resultado)
```

Si una función no tiene `return`, devuelve `None` por defecto:

```python
def sin_retorno():
    print("Esta función no devuelve nada")

print(sin_retorno())
```

## 4.5 Argumentos Variables en Funciones

### 4.5.1 `*args`: Múltiples argumentos posicionales

Cuando no sabemos cuántos argumentos se pasarán a una función, podemos usar `*args` para recibir varios valores.

```python
def sumar_todos(*numeros):
    return sum(numeros)

print(sumar_todos(1, 2, 3, 4, 5))
```

📌 ***args recibe los valores como una tupla**.


### 4.5.2 `**kwargs`: Múltiples argumentos con claves

`**kwargs` permite pasar múltiples argumentos con nombres clave-valor.

```python
def mostrar_info(**datos):
    for clave, valor in datos.items():
        print(f"{clave}: {valor}")

mostrar_info(nombre="Ana", edad=25, ciudad="Santiago")
```

📌 ****kwargs recibe los valores como un diccionario**.

## 4.6 Funciones Anidadas

Las funciones pueden definirse dentro de otras funciones.

```python
def externa():
    print("Esta es la función externa.")

    def interna():
        print("Esta es la función interna.")

    interna()

externa()
```

## 4.7 Funciones Lambda (Funciones Anónimas)

Las **funciones lambda** son expresiones cortas sin nombre que permiten definir funciones en una sola línea.

```python
doblar = lambda x: x * 2
print(doblar(5))
```

Otro ejemplo con dos parámetros:

```python
multiplicar = lambda a, b: a * b
print(multiplicar(3, 4))
```

📌 **Se usan comúnmente en funciones como `map()`, `filter()`, y `sorted()`**.

## 4.8 Alcance de Variables en Funciones

El **ámbito** de una variable determina dónde puede ser utilizada.


### 4.8.1 Variables Locales y Globales
-	**Variable local**: Definida dentro de una función y solo accesible dentro de ella.
-	**Variable global**: Definida fuera de cualquier función y accesible desde cualquier parte del programa.

```python
nombre_global = "Python"

def funcion():
    nombre_local = "Curso"
    print(nombre_global)  # se puede acceder a la variable global
    print(nombre_local)  # solo accesible dentro de la función

funcion()
```

```python
# Error: la variable no está definida fuera de la función
print(nombre_local)
```

📌 **Evita modificar variables globales dentro de funciones, ya que puede causar errores difíciles de depurar**.

## 4.9 Funciones Recursivas

Una **función recursiva** es aquella que se llama a sí misma para resolver problemas repetitivos.

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```
